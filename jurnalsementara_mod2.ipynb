```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def show(img, title=""):
    plt.imshow(img, cmap='gray')  
    plt.title(title)              
    plt.axis('on')    
```

```python
img1 = plt.imread("Assets/Bunga.png")
img2 = plt.imread("Assets/Langit.png")
img3 = plt.imread("Assets/Meledak.png")
img4 = plt.imread("Assets/Windut.png")
```

```python
plt.figure(figsize=(6,6))
show(img3)
plt.show()
```

```python
def slicing(gambar, a, b, mode='H'): 
    return gambar[a:b, :] if mode == 'H' else gambar[:, a:b]
def crop(gambar, x1, y1, x2, y2): 
    return gambar[y1:y2, x1:x2]

potongan1 = slicing(img3, 0, 350, mode='V')
potongan1 = crop(potongan1, 0, 0, 350, 350)
potongan2 = slicing(img3, 350, 700, mode='V')
potongan2 = crop(potongan2, 0, 0, 350, 350)
potongan3 = slicing(img3, 350, 700, mode='H')
potongan3 = crop(potongan3, 0, 0, 350, 350)
potongan4 = slicing(img3, 350, 700, mode='H')
potongan4 = crop(potongan4, 350, 0, 700, 700)

plt.figure(figsize=(10,20))
plt.subplot(1, 2, 1)
show(potongan1, "Part 1")
plt.subplot(1, 2, 2)
show(potongan2, "Part 2")
plt.subplot(2, 2, 3)
show(potongan3, "Part 3")
plt.subplot(2, 2, 4)
show(potongan4, "Part 4")
```

```python
def fungsi_dua_b(citra_1, citra_2): 
    c1, c2 = np.array(citra_1), np.array(citra_2) 
    tinggi = max(c1.shape[0], c2.shape[0]) 
    lebar = c1.shape[1] + c2.shape[1] 
    hasil = np.zeros((tinggi, lebar)).astype(int) 
    hasil[0:c1.shape[0], 0:c1.shape[1]] = c1 
    hasil[0:c2.shape[0], c1.shape[1]:] = c2 

def fungsi_dua_c(citra_1, citra_2): 
    c1, c2 = np.array(citra_1), np.array(citra_2) 
    tinggi = c1.shape[0] + c2.shape[0] 
    lebar = max(c1.shape[1], c2.shape[1]) 
    hasil = np.zeros((tinggi, lebar)).astype(int) 
    hasil[0:c1.shape[0], 0:c1.shape[1]] = c1 
    hasil[c1.shape[0]:, 0:c2.shape[1]] = c2 
    return hasil 

hasil = fungsi_dua_c(img4, img3)
plt.figure(figsize=(5,10))
show(hasil, "Hasil Fungsi Dua C")
```
